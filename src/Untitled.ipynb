```python
%matplotlib inline
from matplotlib import pyplot as plt

import numpy as np
from scipy.misc import imread, imresize
```

```python
hr = imread('../data/General-100/train/im_44.bmp')[160:200, 750:150, :]
lr = imresize(hr, (hr.shape[0] // 2, hr.shape[1] // 2), interp='bicubic')
bicubic = imresize(lr, (lr.shape[0] * 2, lr.shape[1] * 2), interp='bicubic')
sr = imread('./im_44.bmp')[150:250, 50:150, :]
```

```python
plt.figure(figsize=(16,16))

plt.subplot(2, 2, 1)
plt.imshow(hr)
plt.title('HR')

plt.subplot(2, 2, 2)
plt.imshow(lr)
plt.title('LR')

plt.subplot(2, 2, 3)
plt.imshow(bicubic)
plt.title('Bicubic')

plt.subplot(2, 2, 4)
plt.imshow(sr)
plt.title('SR')
```
