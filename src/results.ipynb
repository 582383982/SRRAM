```python
import numpy as np
from scipy.misc import imread, imresize

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
sf = 3
hr = imread('../data/General-100/train/im_51.bmp')[160:240, 70:150, :]
lr = imresize(hr, (hr.shape[0] // sf, hr.shape[1] // sf), interp='bicubic')
bi = imresize(lr, (lr.shape[0] * sf, lr.shape[1] * sf), interp='bicubic')
sr = imread('./im_51.bmp')[160:240, 70:150, :]
```

```python
plt.figure(figsize=(16, 16))

plt.subplot(2,2,1)
plt.imshow(hr)

plt.subplot(2,2,2)
plt.imshow(lr)

plt.subplot(2,2,3)
plt.imshow(bi)

plt.subplot(2,2,4)
plt.imshow(sr)
```

```python
sf = 3
x, y = 150, 50
width = 150
hr = imread('../data/General-100/train/im_44.bmp')

plt.figure(figsize=(16, 16))
plt.imshow(hr)

hr = hr[x:x+width, y:y+width, :]
lr = imresize(hr, (hr.shape[0] // sf, hr.shape[1] // sf), interp='bicubic')
bi = imresize(lr, (lr.shape[0] * sf, lr.shape[1] * sf), interp='bicubic')
sr = imread('./im_44.bmp')[x:x+width, y:y+width, :]

plt.figure(figsize=(16, 16))
plt.subplot(2,2,1)
plt.imshow(hr)
plt.title('HR')

plt.subplot(2,2,2)
plt.imshow(lr)
plt.title('LR')

plt.subplot(2,2,3)
plt.imshow(bi)
plt.title('Bicubic')

plt.subplot(2,2,4)
plt.imshow(sr)
plt.title('SR')
```
